```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import seaborn as sns

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
%matplotlib inline
#dict_ = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'], 'Min': [-1000]*12, 'Max': [0]*12}
data = pd.DataFrame(columns = ['Date', 'Min', 'Max', 'Month'])
```

```python
data['Date'] = df.Date.unique()
data = data.sort_values(by = 'Date')
data.index = range(data.shape[0])
```

```python
data.head()
```

```python
#finding min/max temp from various station readings for every day
for i in range(data.shape[0]):
    date = data.Date[i]
    temps = df[df['Date'] == date]['Data_Value'].values
    min_temp= min(temps)
    max_temp = max(temps)
    data.Min[i] = min_temp/10
    data.Max[i] = max_temp/10
    data['Month'][i] = pd.to_datetime(data.Date[i]).month
```

```python
#only compare values till 2015
for i in range(365, data[data['Date'] == '2015-01-01'].index[0]):
    idx = pd.to_datetime(data.Date[i]).dayofyear - 1
    data.Min[idx] = min(data.Min[idx], data.Min[i])
    data.Max[idx] = max(data.Max[idx], data.Max[i])
```

```python
data_2015 = data[data[data['Date'] == '2015-01-01'].index[0]:]
data_2015
```

```python
x = []
y = []
for i in range(365):
    if data_2015.Min[i] < data.Min[i]:
        x.append(i)
        y.append(data_2015.Min[i])
    if data_2015.Max[i] > data.Max[i]:
        x.append(i)
        y.append(data_2015.Max[i])
```

```python
groups = data.groupby('Month')
xticks = []
for i in range(12):
    xticks.append(groups.count()['Date'][:i].sum())
```

```python
import numpy as np
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

plt.figure(edgecolor='lightgrey', figsize=(10,8))

plt.plot( data.Min, linewidth = 0.5);
plt.plot( data.Max, linewidth = 0.5);

plt.title('Max and Min Temperature($^{\circ}$C) over 2005-14 with \n record breaking temperatures in 2015');
for spine in plt.gca().spines:
    plt.gca().spines[spine].set_visible(False);
    
plt.gca().spines['bottom'].set_visible(True);
plt.gca().grid(False);
plt.xticks(xticks, labels);
plt.scatter(x, y, c = 'r', s = 10, label  = '2015 Temperatures');
plt.legend();
plt.fill_between(data.index, [float(x1) for x1 in data.Min], [float(y1) for y1 in data.Max], facecolor = 'grey', alpha = 0.15);
```
